```python
import numpy as np
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
import pandas as pd # Import pandas for one-hot encoding


# Re-create 'a' with one-hot encoding to handle categorical data
categorical_cols = d.columns[3:] # Select columns from index 3 onwards
a_df = d[categorical_cols]
a_encoded = pd.get_dummies(a_df, drop_first=True) # Apply one-hot encoding
a = a_encoded.values # Convert to numpy array

# Agglomerative Clustering
model = AgglomerativeClustering(
    n_clusters=4,
    metric='euclidean',
    linkage='ward'
)

labels = model.fit_predict(a)

print("Cluster Labels:")
print(labels)

# Visualization
# Note: Visualizing high-dimensional data (after one-hot encoding) with just the first two components
# might not accurately represent the clusters. For better visualization, consider dimensionality
# reduction techniques like PCA or t-SNE before plotting.
if a.shape[1] >= 2:
    plt.scatter(a[:,0], a[:,1], c=labels, cmap='rainbow')
    plt.title("Agglomerative Clustering")
    plt.show()
else:
    print("Cannot visualize clusters in 2D as the data has less than 2 dimensions after encoding.")
```

```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


def divisive_clustering(data, n_clusters_target):
    clusters = [data] # Initialize clusters with the entire dataset as the first cluster

    # Continue splitting until the target number of clusters is reached
    while len(clusters) < n_clusters_target:
        # Find the largest cluster by the number of data points
        largest_cluster_index = -1
        max_size = -1
        for i, cluster in enumerate(clusters):
            if len(cluster) > max_size:
                max_size = len(cluster)
                largest_cluster_index = i

        # Handle cases where no cluster is found (e.g., empty list) or if splitting a tiny cluster
        if largest_cluster_index == -1 or max_size == 0:
            break # Cannot proceed if no valid cluster to split

        # Remove the largest cluster from the list by its index
        cluster_to_split = clusters.pop(largest_cluster_index)

        # Ensure the cluster has enough samples to be split by KMeans (min 2 samples for n_clusters=2)
        if len(cluster_to_split) < 2:
            # If a cluster is too small to split, add it back and stop further splitting of this path
            clusters.append(cluster_to_split)
            # If we cannot make more clusters to reach n_clusters_target, we must stop
            break

        # Perform K-means with k=2 to divide the largest cluster
        kmeans = KMeans(n_clusters=2, random_state=42, n_init='auto') # Added n_init for KMeans warning
        labels = kmeans.fit_predict(cluster_to_split)

        # Split the cluster into two new sub-clusters based on K-means labels
        new_cluster_1 = cluster_to_split[labels == 0]
        new_cluster_2 = cluster_to_split[labels == 1]

        # Add the two new sub-clusters back to the list
        clusters.append(new_cluster_1)
        clusters.append(new_cluster_2)

    return clusters

clusters = divisive_clustering(a, 3)
for i, cluster in enumerate(clusters):
    print(f"\nCluster {i+1}")
    print(cluster)

# Visualization
colors = ['red', 'blue', 'green', 'purple']

# Check if data has at least 2 dimensions for 2D scatter plot
if a.shape[1] >= 2:
    for i, cluster in enumerate(clusters):
        # Ensure cluster is not empty before plotting
        if cluster.shape[0] > 0:
            plt.scatter(cluster[:,0], cluster[:,1],
                        color=colors[i % len(colors)],
                        label=f'Cluster {i+1}')

    plt.legend()
    plt.title("Divisive Clustering")
    plt.show()
else:
    print("Cannot visualize clusters in 2D as the data has less than 2 dimensions after encoding.")
```
